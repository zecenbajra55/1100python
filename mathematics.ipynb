# kurtosis refers to the degree of presence of outliers in the distribution . kurtosis is a statistical measure whether the date si heavy or light tailed in a normal distribution

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(40)
data=np.random.normal(loc=5,scale=1,size=3000)
```

```python
sns.histplot(data,kde=True)
```

# kurtosis type
* 1. mesokurtic ~ normal distribution kurtois ~ 3

```python
from scipy.stats import kurtosis
sns.histplot(data,kde=True)
kurt=kurtosis(data, fisher=False)
plt.title(f'kurtosis={round(kurt,2)}')
plt.show()
```

* 2. platykurtic distributions have kurtosis < 3
** uniform distribution

```python
np.random.seed(40)
data=np.random.uniform(low=2,high=6,size=30000)
sns.histplot(data,kde=True)
kurt=kurtosis(data, fisher=False)
plt.title(f'kurtosis={round(kurt,2)}')
plt.show()
```

* no outlier in platykurtic distribution

* 3. leptokurtic distribution is fat-tailed meaning that there are alot of outliers kurtosis > 3. it is also called laplace distribution

```python
np.random.seed(40)
data=np.random.laplace(size=3000)
sns.histplot(data,kde=True)
kurt=kurtosis(data, fisher=False)
plt.title(f'kurtosis={round(kurt,2)}')
plt.show()
```
