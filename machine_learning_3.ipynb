# encoding categorical variables

```python
import pandas as pd
df = pd.DataFrame({'education':['bachelors','phd','phd','masters','masters','bachelors']})
```

```python
df
```

# one hot encoding

```python
pd.get_dummies(columns = ['education'],data =df)
```

```python
# remove multi colinarity by dropping first column
pd.get_dummies(columns=['education'],data=df,drop_first = True)
```

# one hot encoding using sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output = False) # sparse_output = False returns numpy array
encoder.fit(df[['education']]) # .fit() method needs data in 2D
encoded=encoder.transform(df[['education']])
encoded
```

```python
encoder.categories_
```

```python
pd.DataFrame(encoded, columns = encoder.categories_)
```

```python
encoder = OneHotEncoder(sparse_output = False,drop ='first') # sparse_output = False returns numpy array
encoder.fit(df[['education']]) # .fit() method needs data in 2D
encoded=encoder.transform(df[['education']])
encoded
```

# label encoder

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(df['education'])  #  .fit() expects 1D
encoder.transform(df['education'])
```

```python
encoder.classes_
```

```python
encoder = LabelEncoder()
encoder.fit_transform(df['education'])
```
