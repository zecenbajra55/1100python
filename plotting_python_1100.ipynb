```python
import matplotlib.pyplot as plt
```

```python
year=['2020','2021','2022','2023']
growth = [3.4,4.7,8,8.9]
```

```python
plt.plot(year,growth,'o-',color='green')
plt.xlabel('year')
plt.ylabel('growth%')
plt.title('yearly growth')
plt.show()
```

# bar plot

```python
gender=['female','male']
count=[80,62]
plt.figure(figsize=(4,2))
plt.bar(x=gender,height=count, color=['cyan','red'])
plt.show()
```

# pichart

```python
gender=['female','male']
count=[80,62]
plt.figure(figsize=(4,2))
plt.pie(x=count,labels=gender, autopct='%1.1f%%',colors=['green','blue'])
plt.show()
```
