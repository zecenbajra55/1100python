```python
about={
    'name':['sisan baniya','shrinkhala khatiwada','cool boy'],
    'social_media':['youtube','instagram','tiktok'],
    'followers':['713K','1M','557.8K']
};
```

```python
df=pd.DataFrame(about)
```

```python
def clean(x):
    if 'K' in x :
        x=x.replace('K','')
        x= float(x) *1000
        return x
    elif "M" in x:
        x=x.replace('M','')
        x=float(x)*1000000
        return x
    return x
```

```python
df['followers']= df['followers'].apply(clean)
```

```python
df.head()
```

```python
# rename column of a dataframe
```

```python
df.rename({'social_media': 'Social Media',
           'name':'Name',
           'followers':'Total_Followers'
           },axis='columns')
```

```python
df # dataframe is not updated
```

```python
df.rename({'social_media': 'Social Media',
           'name':'Name',
           'followers':'Total_Followers'
           },axis='columns',inplace=True)      # dataframe mai updata garxa inplace= True le
```
